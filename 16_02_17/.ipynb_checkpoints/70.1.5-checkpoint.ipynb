70. N(23.56, 0.27)

```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = 23.56
sigma = 0.27

N = 10 ** 6
num_of_samples = 10

samples = [sps.norm.rvs(loc=a, scale=sigma, size=N) for i in range(num_of_samples)]
for i in range(num_of_samples):
    plt.hist(samples[i], alpha = 1 / num_of_samples * 2)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Samples")
fig = plt.gcf()

```

```python
def stat_CLT(X):
    return (len(X) / sigma)**(1/2) * (np.mean(X) -  a) 
```

```python
Y = np.arange(0, N // 100, 5)
for i in range(num_of_samples):
    stat = [stat_CLT(samples[i][:j]) for j in Y]
    plt.plot(Y, stat, alpha = 1 / num_of_samples * 2)
plt.xlabel("n")
plt.ylabel("Value")
plt.title("Statistics for different samples from 0 to n element")
fig = plt.gcf()
```

```python
CLT_stat = []
Y = np.arange(0, N , 10)
for i in range(num_of_samples):
    stat = [stat_CLT(samples[i][:j]) for j in Y]
    CLT_stat.append(stat)
    plt.plot(Y, CLT_stat[i], alpha = 1 / num_of_samples * 2)
plt.xlabel("n")
plt.ylabel("Value")
plt.title("Statistics for different samples from 0 to n element")
fig = plt.gcf()
#поведение статистики от n(на большем диапазоне, чем на предыдущем графике)
```

```python
Y = np.arange(0, N - 101 , 100)
for i in range(num_of_samples):
    stat = [np.mean(CLT_stat[i][j:]) for j in Y]
    plt.plot(Y, stat, alpha = 1 / num_of_samples * 2)
plt.title("Mean for statistics for different samples from n to N element")
plt.show()
# поведение мат.ожидания хвоста статистики(от n до N)
```

```python
Y = np.arange(0, N - 1000 , 100)
for i in range(num_of_samples):
    stat = [np.var(CLT_stat[i][j:]) for j in Y]
    plt.plot(Y, stat, alpha = 1 / num_of_samples * 2)
plt.title("Variance for statistics for different samples from n to N element")
plt.show()
# поведение дисперсии хвоста статистики(от n до N)
```

```python
print(len(CLT_stat[1]))
plt.figure(figsize=(20,15))
for i in range(num_of_samples - 1):
    plt.subplot(3,3,i + 1)
    plt.hist(CLT_stat[i][99000:], alpha = 1 / num_of_samples * 2)
    plt.xlabel("Value")
    plt.ylabel("Frequency")
plt.title("Samples")
plt.show()
# гистограммы для хвостов получившихся статистик
```

Статистика $T_n$ ведёт себя в большинстве случаев близко к нормальному распределению с мат.ожиданием $\sim 0$ и выборочной дисперсией близкой к нулю, откуда следует, что ЦПТ применима, но лишь с некоторыми ограничениями для произвольной выборки данного размера(например, нельзя строго утверждать о классификации данных статистик).
