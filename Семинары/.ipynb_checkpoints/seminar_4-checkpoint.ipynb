```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
from statsmodels.sandbox.stats.multicomp import multipletests

%matplotlib inline
```

## Множественная проверка гипотез

Все процедуры реализуются простой функцией. Полная документация по ссылке.

<a href=http://statsmodels.sourceforge.net/devel/generated/statsmodels.sandbox.stats.multicomp.multipletests.html>`multipletests`</a>`
(pvals, alpha=0.05, method='hs', is_sorted=False, returnsorted=False)`

-----

`pvals` --- набор p-value

`alpha` --- контроль FWER или FDR

`method` --- метод (см. документацию)

`is_sorted` --- если true, то `pvals` должны быть уже отсортированы

-----

Возвращает: `reject, pvals_corrected, alphacSidak, alphacBonf`

`reject` --- отвергаемые гипотезы

`pvals_corrected` --- корректировка p-value. Смысл - их можно сравнивать непосредственно с `alpha`.

-----

Просто пример

```python
multipletests([0.01, 0.02, 0.05, 0.07])
```

Как модифицированные pvalue связаны с отклонением гипотез

```python
reject, pvals_corrected = multipletests([0.016, 0.05, 0.07])[:2]
print(reject)
print(pvals_corrected)

reject, pvals_corrected = multipletests([0.017, 0.05, 0.07])[:2]
print(reject)
print(pvals_corrected)
```

Метод Бонферрони отвергает больше гипотез:

```python
reject, pvals_corrected = multipletests([0.001, 0.002, 0.012, 0.02, 
                                         0.05, 0.07])[:2]
print(reject)
print(pvals_corrected)

reject, pvals_corrected = multipletests([0.001, 0.002, 0.012, 0.02, 
                                         0.05, 0.07], 
                                        method='bonferroni')[:2]
print(reject)
print(pvals_corrected)
```

**Вывод:** Оно работает, клево! <font size="6" color="red">&#12483;</font>

### Пример: поиск экстрасенсов

#### Бернуллиевская выборка

$X_1, ..., X_n \sim Bern(\theta)$

$H_0: \theta = 1/2$

$H_1: \theta > 1/2$

$\left\{\sum\limits_{i=1}^{10} X_i \geqslant c_{\alpha}\right\}$ --- критерий, где $c_{\alpha}$ --- $(1-\alpha)$-квантиль $Bin(10, 1/2)$

```python
alpha = 0.05
c_alpha = sps.binom(n=10, p=0.5).ppf(1 - alpha)
print('Критерий: {sample_sum > %.d}' % c_alpha)
```

Почему в итоге знак строгий (`sf = 1 - cdf`):

```python
print('%.3f, %.3f' % (sps.binom(n=10, p=0.5).sf(7.5), 
                      sps.binom(n=10, p=0.5).sf(8.5)))
```

Берем 1000 человек (генерируем сразу количество верно отгаданных из биномиального распределения) и проверяем 1000 гипотез. Смотрим, во скольких случаях верную нулевую гипотезу отклоняем, то есть когда количество верно отгаданных превышает критическое значение.

```python
sample = sps.binom(n=10, p=0.5).rvs(size=1000)
print((sample > 8).sum())
```

Для каждой гипотезы считаем pvalue. Поправка `-0.5` для того, чтобы не было проблем со строгими/нестрогими знаками. Красной линией отмечен уровень значимости $\alpha = 0.05$.

```python
pvalues = sps.binom(n=10, p=0.5).sf(sample - 0.5)

plt.figure(figsize=(12, 6))
plt.scatter(np.arange(1, 1001), np.sort(pvalues))
plt.hlines(0.05, -100, 1100, color='red')
plt.xlim((-30, 1030))
plt.grid(ls=':')
plt.xlabel('rank of p-value', fontsize=16)
plt.ylabel('p-value', fontsize=16)
plt.show()
```

Применим множественную проверку гипотез (метод Шидака-Холма - применим, поскольку все выборки независимы)

```python
reject, pvals_corrected = multipletests(pvalues)[:2]
print(reject.sum())

plt.figure(figsize=(12, 6))
plt.scatter(np.arange(1, 1001), np.sort(pvals_corrected))
plt.hlines(0.05, -100, 1100, color='red')
plt.xlim((-30, 1030))
plt.grid(ls=':')
plt.xlabel('rank of p-value', fontsize=16)
plt.ylabel('p-value', fontsize=16)
plt.show()
```

Теперь уже ни одна гипотеза не была отвергнута.

### Модельный эксперимент

$X_1^{(j)}, ..., X_{20}^{(j)} \sim \mathcal{N}(0, 1),\ \ \ \ \ j = 1,..., 150$

$X_1^{(j)}, ..., X_{20}^{(j)} \sim \mathcal{N}(1, 1),\ \ \ \ \ j = 151,..., 200$

$H_j: \mathsf{E} X_i^{(j)} = 0$

$\widehat{H}_j: \mathsf{E} X_i^{(j)} \not= 0$

Используется одновыборочный $t$-критерий (Стьюдента).

Статистика: $$T(X) = \sqrt{n}\frac{\overline{X}}{\sqrt{S^2}} \sim T_{n-1}\ \ \ \text{(при $H_0$)}$$

Критерий: $\left\{\left|T(X)\right| \geqslant c_{\alpha}\right\}$ --- критерий, где $c_{\alpha}$ --- $(1-\alpha/2)$-квантиль $T_{n-1}$

Генерация выборок

```python
n = 100
m = 200
means = np.zeros((m, n))
means[150:] = 1
sample = sps.norm(loc=means).rvs(size=(m, n))
```

Изобретем велосипед, посчитав все pvalue руками. В данном случае так эффективнее, поскольку иначе при использовании готовой функции пришлось бы перебирать в цикле вызовы этой функции для каждой выборки. Здесь `t` реализует распределение Стьюдента, а `df` - число степеней свободы. Домножение на 2 поскольку критерий двусторонний.

```python
T = np.sqrt(n) * sample.mean(axis=1) / sample.std(axis=1)
pvalues = 2 * sps.t(df=n-1).sf(np.abs(T))
```

Функция, которая рисует график rank-pvalue.

```python
def draw_sorted_pvalues(pvalues, truth=None, figsize=(12, 6), title=''):
    m = len(pvalues)
    order = np.argsort(pvalues)
    pvalues = pvalues[order]
    if truth is not None:
        truth = truth[order]
    
    plt.figure(figsize=figsize)
    x = np.arange(len(pvalues))
    
    if truth is None: 
        plt.scatter(x, pvalues)
    else:
        plt.scatter(x[truth], pvalues[truth], c='blue')
        plt.scatter(x[~truth], pvalues[~truth], c='red')
        
    plt.hlines(0.05, -100, m * 1.1, color='red')
    plt.xlim((-m / 50, m * 1.05))
    plt.grid(ls=':')
    plt.xlabel('rank of p-value', fontsize=16)
    plt.ylabel('p-value', fontsize=16)
    plt.title(title, fontsize=16)
    plt.show()
```

График обычных (немодифицированных) pvalue. Синие и красные pvalue соответствуют верным и неверным нулевым гипотезам.

```python
draw_sorted_pvalues(pvalues, truth=np.arange(200)<150)
```

Графики pvalue после корректировки. Все процедуры множественной проверки гипотез в данном случае применимы, поскольку все выборки независимы, соответственно независимы статистики критериев. На практике это не всегда так.

```python
for method in ['hs', 'bonferroni', 'sidak', 'holm', 'fdr_bh', 'fdr_by']:
    pvals_corrected = multipletests(pvalues, method=method)[1]
    draw_sorted_pvalues(pvals_corrected, truth=np.arange(200)<150, title=method)
```

## Критерии случайности

Функция, реализующая *критерий инверсий*. Возвращает статистику T и pvalue.

```python
def inversions_criterion(sample):
    x = np.array(sample)
    n = len(x)
    
    upper_triangular = np.arange(n).reshape((-1, 1)) < np.arange(n).reshape((1, -1))
    x_upper_triangular = x.reshape((-1, 1)) > x.reshape((1, -1))
    I = (x_upper_triangular * upper_triangular).sum()
    
    mean = n * (n - 1) / 4
    var = (2 * n ** 3 + 3 * n ** 2 - 5 * n) / 72
    T = (I - mean) / np.sqrt(var)
    return T, 2 * sps.norm.sf(np.abs(T))
```

Простой пример

```python
inversions_criterion([1, 2, 3, 4, 5, 6.5, 6, 7.5, 7, 8])
```

Передадим критерию текущие баллы студентов в 3 группе МСПС, порядок согласно таблице, то есть по алфавиту. Случайно ли распределяются баллы? Видимо, да. Обратное означало бы наличие зависимости между фамилией по алфавиту и количеством полученных баллов.

```python
inversions_criterion([5, 11, 9, 7, 9.5, 2, 7.5, 11, 5, 17, 8.5, 
                      8, 13, 11.5, 11.5, 8, 11, 10, 14, 10, 0])
```

Теперь передадим баллы студентов всех трех групп за первое ДЗ. Порядок согласно таблице, то есть сначала сортировка по группам, затем по алфавиту. Случайно ли произошло деление на группы по МСПС? Видимо, нет - pvalue мало.

```python
inversions_criterion([6, 8, 1.5, 0, 0, 4, 10, 4.5, 0, 12, 0, 2.5, 
                      4, 7.5, 3.5, 6, 8, 0, 0, 9.5, 0, 6.5, 3, 0, 
                      0, 7.5, 12, 11.5, 4, 2, 11.5, 0, 11.5, 0, 8,
                      7.5, 0, 0, 0, 0, 0, 0, 9.5, 5, 9.5, 5.5, 12, 
                      2, 5, 5, 3, 10, 6, 6, 7.5, 2, 6.5, 9, 3, 10, 
                      8.5, 8, 10, 10.5, 11.5, 5, 9, 6, 9, 8, 0])
```

Оставим только баллы первых двух групп. Случайно ли произошло деление студентов между 1 и 2 группой. Видимо, да.

```python
inversions_criterion([6, 8, 1.5, 0, 0, 4, 10, 4.5, 0, 12, 0, 2.5, 
                      4, 7.5, 3.5, 6, 8, 0, 0, 9.5, 0, 6.5, 3, 0, 
                      0, 7.5, 12, 11.5, 4, 2, 11.5, 0, 11.5, 0, 8,
                      7.5, 0, 0, 0, 0, 0, 0, 9.5, 5, 9.5, 5.5, 12, 
                      2, 5, 5])
```
