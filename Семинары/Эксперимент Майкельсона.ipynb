```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt

from statsmodels.distributions.empirical_distribution import ECDF
from sklearn.neighbors import KernelDensity
import statsmodels.api as sm

%matplotlib inline
```

Данные классического эксперимента Майкельсона по измерению скорости света с помощью вращающегося зеркала, 100 наблюдений:


```python
speed = np.loadtxt('speed.txt')

plt.figure()
plt.scatter(speed, np.zeros_like(speed) - 0.001, alpha=0.3)
plt.hist(speed, bins=15, normed=True)
plt.ylim((-0.002, 0.007))
plt.xlabel('speed')
plt.show()
```

```python
mean = speed.mean()
std = speed.std()
grid = np.linspace(speed.min(), speed.max(), 1000)

plt.figure(figsize=(12, 5))
plt.plot(grid, sps.norm(mean, std).cdf(grid), label='cdf $\mathcal{N}' + '(%d, %d)$' % (mean, std**2))
plt.plot(grid, ECDF(speed)(grid), label='ecdf')
plt.legend(fontsize=20)
plt.xlabel('speed', fontsize=20)
plt.show()

plt.figure(figsize=(12, 5))
plt.plot(grid, sps.norm(mean, std).pdf(grid), label='pdf $\mathcal{N}' + '(%d, %d)$' % (mean, std**2))
plt.plot(grid, 
         np.exp(KernelDensity(bandwidth=25)
                .fit(speed[:, np.newaxis])
                .score_samples(grid[:, np.newaxis])), 
         label='kde')
plt.legend(fontsize=20)
plt.xlabel('speed', fontsize=20)
plt.show()
```

```python
sps.kstest(speed, cdf='norm')
```

```python
sps.kstest(speed, cdf=sps.norm(mean, std).cdf)
```

```python
sps.anderson(speed)
```

```python
sm.qqplot(speed, line='r')
plt.show()
```

```python
sps.shapiro(speed)
```
