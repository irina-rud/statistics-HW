#### 2.
Плотность распределения Коши и плотность стандартного нормального распределения.

```python
grid = np.linspace(-7, 7, 1000)

plt.figure(figsize=(10, 4))
plt.plot(grid, sps.cauchy.pdf(grid), color='red', label='Cauchy')
plt.plot(grid, sps.norm.pdf(grid), color='blue', label='normal')
plt.legend(fontsize=16)
plt.grid(ls=':')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

Сравнение поведения оценок $\overline{X}$ и $\widehat{\mu}$ для модели распределения Коши при увеличении размера выборки.

```python
size = 500
time = np.arange(1, size + 1)
sample = sps.cauchy.rvs(size=size)
avg_estimation = np.cumsum(sample) / time
median_estimation = np.array(list([np.median(sample[:n]) for n in time]))

plt.figure(figsize=(15, 5))
plt.plot(time, avg_estimation, color='red', label='$\overline{X}$')
plt.plot(time, median_estimation, color='blue', label='$\widehat{\mu}$')
plt.legend(fontsize=16)
plt.grid(ls=':')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

#### 3.
Плотность распределения $\Gamma(1, 2)$.

```python
alpha, beta = 1, 2
grid = np.linspace(-0.5, 7, 1000)

plt.figure(figsize=(10, 4))
plt.plot(grid, sps.gamma(a=beta, scale=1/alpha).pdf(grid), color='red', label='density')
plt.legend(fontsize=16)
plt.grid(ls=':')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

Сравнение ОМП и байесовской оценки в модели $Exp(\theta)$ при $\theta_0=2$. В качестве априорного распределения взято $\Gamma(1, 2)$.

```python
size = 50
time = np.arange(1, size + 1)
sample = sps.expon(scale=1/2).rvs(size=size)
ml_estimation = time / np.cumsum(sample)
bayes_estimation = (beta + time) / (alpha + np.cumsum(sample))

plt.figure(figsize=(15, 5))
plt.plot(time, ml_estimation, color='red', label='ML')
plt.plot(time, bayes_estimation, color='blue', label='Bayes')
plt.legend(fontsize=16)
plt.grid(ls=':')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

Сравнение ОМП и байесовской оценки в модели $Exp(\theta)$ при $\theta_0=2$. В качестве априорного распределения взято $\Gamma(1, 20)$.

```python
alpha, beta = 1, 20
size = 50
time = np.arange(1, size + 1)
sample = sps.expon(scale=1/2).rvs(size=size)
ml_estimation = time / np.cumsum(sample)
bayes_estimation = (beta + time) / (alpha + np.cumsum(sample))

plt.figure(figsize=(15, 5))
plt.plot(time, ml_estimation, color='red', label='ML')
plt.plot(time, bayes_estimation, color='blue', label='Bayes')
plt.legend(fontsize=16)
plt.grid(ls=':')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

#### 4.
График доверительных интервалов для модели $\mathcal{N}(\theta, 1)$ в зависимости от времени..

```python
time = np.arange(1, 201)
sample = sps.norm.rvs(size=200)
means = sample.cumsum() / np.arange(1, 201)

plt.figure(figsize=(15, 8))
plt.scatter(time, sample, alpha=0.5, s=40, label='sample')
plt.plot(time, means, color='red', linewidth=2.5, label='sample mean $\overline{X}$')
plt.fill_between(time, means + 2 / np.sqrt(time), means - 2 / np.sqrt(time), alpha=0.15)
plt.legend()
plt.xlim((1, 200))
plt.ylim((-3, 3))
plt.xlabel('Time')
plt.title('Sample by time')
plt.grid()
plt.show()
```
