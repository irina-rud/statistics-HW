```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sps
import pandas as pd
from statsmodels.sandbox.stats.multicomp import multipletests


%matplotlib inline

```

```python
data = pd.read_csv('tic-tac-toe.data', header = None, sep=',')
data.head()
```

```python
N = 10
def make_num(st):
    if (st.strip() == 'x'):
        return 1
    elif (st.strip() == 'o'):
        return 0
    elif (st.strip() == 'b'):
        return 2
    elif (st.strip() == 'positive'):
        return 3
    elif (st.strip() == 'negative'):
        return 4
    else:
        print(st)
        return None

x = []
for i in range(N):
    x.append(np.array(list(map(make_num,data.values[:,i]))))
x = np.array(x)

```

```python
print(x[1])

```

```python
def normility(samples, alpha = 0.05, method='bonferroni'):
    p_val = np.zeros(N * N)
    for i in range(N):
        for j in range(N):
            p_val[i + N * j] = sps.kendalltau(x[i], x[j])[1]
    return multipletests(p_val, alpha=alpha, method=method)

for method in ['bonferroni', 'holm']:
    reject, pvals_corrected =  normility(np.array(x), alpha=0.05, method=method)[:2]
    print("method %s \n" % (method))
    for i in range(N):
        for j in range(i, N):
            if (i != j):
                print("гипотеза о независимости (%d, %d): " % (i, j) +str(reject[i + N * j]))
    print()

        
        
```
