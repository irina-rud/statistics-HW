```r
x <- rnorm(n = 500, mean = 1, sd = 1)
```

```r
a <- mean(x)
a
```

```r
v <- var(x)
v
```

```r
b1 <- qnorm(.025, mean = 1, sd = 1)
b1
```

```r
b2 <- qnorm(.975, mean = 1, sd = 1)
b2
```

```r
hist(x, freq = FALSE, col = 3, nclass = 50)
abline(v=b1)
abline(v=b2)
```

```r
s <- 0
for (i in x) {
    if ((i < b1) | (i > b2)){
        s <- s + 1        
    }
}
s
```
