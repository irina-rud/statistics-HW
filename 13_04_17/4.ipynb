```r
all <- read.table("housing.data")
sample <- all[, 14]
all
```

```r
plot(all$V14 ~ all$V5)
model <- lm(formula = V14 ~ V5, 
            data = all)
model$coefficients
summary(model)
x <- seq(from = 0, to = 1, by = 0.01)
lines(x, model$coefficients[2] * x + model$coefficients[1], 
      col = "red", lwd = 2)
```

```r
rest <- model$coefficients[2] * all$V5 + model$coefficients[1] - all$V14
qqnorm(rest)
qqline(rest)
```

Нет, это не похоже на нормальное распределение, больше похоже на Коши
